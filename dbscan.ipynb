```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN,KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X, y_true = make_moons(n_samples=500, noise=0.05 , random_state=42)
```

```python
df = pd.DataFrame(X, columns=['Feature_1', 'Feature_2'])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
kmeans_final = KMeans(n_clusters = 2, random_state = 42)
```

```python
kmeans_labels = kmeans_final.fit_predict(X_scaled)
```

```python
df['kmeans_cluster'] = kmeans_labels
```

```python
sns.scatterplot(x=df['Feature_1'], y=df['Feature_2'], hue=df['kmeans_cluster'], palette='viridis')
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
```

```python
df['dbscan_labels'] = dbscan_labels
```

```python
sns.scatterplot(x=df['Feature_1'], y=df['Feature_2'], hue=df['dbscan_labels'], palette='tab10')
```
