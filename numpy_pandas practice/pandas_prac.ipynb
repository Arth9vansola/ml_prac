```python

keep = {
  'name': ['Alice', 'Bob', 'Charlie'],
  'age': [25, 30, 35],
  'city': ['New York', 'Los Angeles', 'Chicago'],
  'salary': [70000, 80000, 90000]
}
df = pd.DataFrame(keep)
df
```

```python
# creating dataframe using list of lists
data_2 = [['Alice', 25, 'New York', 70000],
          ['Bob', 30, 'Los Angeles', 80000],        
          ['Charlie', 35, 'Chicago', 90000]]              

df = pd.DataFrame(data_2, columns=['name', 'age', 'city', 'salary'])
df
```
