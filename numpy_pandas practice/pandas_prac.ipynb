```python
# finding missing data
data = {
  'A':[1,2,np.nan,4,5],
  'B':[np.nan,2,3,4,5],
  'C':[1,2,3,np.nan,np.nan],
  'D':[1,np.nan,np.nan,np.nan,5]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
# removing missing data
df.dropna()
```

```python
df.dropna(thresh=3) # thresh: indicate there is in row atleast these much or more value should be non null, row with less value than thresh is removed
```

```python
# filling the missing data
df.fillna(0)
```

```python
values = {'A':10,'B':20,'C':30,'D':40}
df.fillna(value=values)
```

```python
# dropna = row wise droping
# fillna = column wise filling
```

```python
df.fillna(df.mean())
```

```python
# merging, joining and concatenation
employee = pd.DataFrame({
  'employee_id':[1,2,3,4,5],
  'name':['avi','arth','arthur','vans','vanzola'],
  'department':['IT','Finance','Gaming','Development','Designing']
})

salaries = pd.DataFrame({
  'employee_id':[1,2,3,6,7],
  'salary':[60000,40000,5000,4622,455],
  'bonus':[1,2,5,6,8]
})

```

```python
employee
```

```python
salaries
```

```python
pd.merge(employee, salaries, on='employee_id',how='outer') #on = common column in both , how = which join you want to use
```

```python
pd.merge(employee, salaries, on='employee_id',how='left')
```

```python
pd.merge(employee, salaries, on='employee_id',how='right')
```

```python
# concatination of 2 dataframes
df1 = pd.DataFrame({
  'A':['A0','A1','A2'],
  'B':['B0','B1','B2'],
  'C':['C0','C1','C2'],
})

df2 = pd.DataFrame({
  'A':['A3','A4','A5'],
  'B':['B3','B4','B5'],
  'C':['C3','C4','C5'],
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df2,df1],ignore_index=True)
```

```python
pd.concat([df1,df2],ignore_index=True, axis=1)
```

```python
# if we want to merge two dataframe based on column then we use merge
# if we want to concat two dataframe based on column or row then we use concat
```

```python
# joining two data frames
df1 = pd.DataFrame({
  'name':['alice','bob','charlie']
}, index=[1,2,3])


df2 = pd.DataFrame({
  'score':[80,60,90]
}, index=[2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2,how='outer')
```
