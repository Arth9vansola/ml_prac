```python
print(arr.min(), arr.max(), arr.sum(), arr.mean(), arr.std(), arr.var())
```

```python
print(np.sum(arr, axis=0)) # column wise sum
print(np.sum(arr, axis=1)) # row wise sum
```

```python
print(np.mean(arr, axis=0)) # column wise min
print(np.mean(arr, axis=1)) # row wise min
```

```python
print(arr.mean())
print(arr.mean(axis=0)) # column wise mean
print(arr.mean(axis=1)) # row wise mean
```

```python
print(arr.argmax()) # index of max value
print(arr.argmin()) # index of min value
```
