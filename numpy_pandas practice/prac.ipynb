```python
import numpy as np

s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])

target = np.arange(1, 10)

sorted_columns = np.sort(s, axis=0) 
for col in sorted_columns.T:        
    if not np.array_equal(col, target):
        print("Not a valid sudoku for column wise verification")
        break
else:
    print("Valid sudoku for column wise verification")


sorted_rows = np.sort(s, axis=1)    
for row in sorted_rows:
    if not np.array_equal(row, target):
        print("Not a valid sudoku for row wise verification")
        break
else:
    print("Valid sudoku for row wise verification")

valid_boxes = True

for i in range(0, 9, 3):
    for j in range(0, 9, 3):

        box = s[i:i+3, j:j+3].flatten()

        if not np.array_equal(np.sort(box), target):
            valid_boxes = False
            break

    if not valid_boxes:
        break

if valid_boxes:
    print("Valid Sudoku Boxes")
else:
    print("Invalid Sudoku Boxes")  
```
