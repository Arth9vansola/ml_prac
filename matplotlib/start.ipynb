```python
import matplotlib.pyplot as plt
```

```python
day = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday']
sales = [200, 100, 250, 5000, 1100]

# plt.bar(day, sales)
plt.plot(day, sales, color='red', marker='o', linestyle='--', linewidth=1, markersize=5, label='Sales')
# plt.bar(day, sales)
# plt.boxplot(sales)
# plt.scatter(day, sales)
# plt.pie(sales, labels=day, autopct='%1.1f%%')
plt.title('Sales per day')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.legend(loc='upper left')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.xlim('monday','friday')
plt.ylim(0,6000)
plt.xticks(['monday', 'tuesday', 'wednesday', 'thursday', 'friday'],['D1','D2','D3','D4','D5'])
plt.show()
```

```python
regions = ['north', 'south', 'east', 'west']
revenue = [1000, 1500, 2000, 2500]
plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['red', 'green', 'blue', 'orange'], startangle=90, shadow=True, explode=(0.1, 0, 0, 0))
plt.show()
```

```python
scores = [50, 20, 55, 80, 93, 45, 55, 65, 75, 85, 95, 40, 30, 20, 10, 100, 35, 25, 15, 5]
plt.hist(scores, bins = 20, color='blue', edgecolor='black')
plt.xlabel('Scores')
plt.ylabel('students')
plt.title('Distribution of Scores')
plt.show()
```

```python
96# scatter plot
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 100, 86, 96, 87, 94, 78, 77, 85, 86]
plt.scatter(x, y, color='red', marker='o', s=100, alpha=1, label='marks vs hours')
plt.title('Relationship between hours studied and scores obtained') 
plt.xlabel('no of hours studied')
plt.ylabel('Score')
plt.legend(loc='upper left')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

```python
# comparison of two classes in terms of scores vs hours studied in a scatter plot

plt.scatter([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6], [99, 86, 87, 88, 100, 86, 96, 87, 94, 78, 77, 85, 86], color='red', marker='o', s=100, alpha=1, label='class A')
plt.scatter([4, 6, 7, 8, 3, 15, 3, 8, 5, 10, 11, 8, 5], [90, 80, 85, 86, 95, 80, 90, 85, 92, 70, 75, 80, 85], color='blue', marker='s', s=100, alpha=1, label='class B')
plt.title('Comparison of Two Classes')
plt.xlabel('no of hours studied')
plt.ylabel('Score')
plt.legend(loc='upper left')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

```python
# subplots using line chart and bar chart

x = [1, 2, 3, 4, 5]
y1 = [2, 50, 5, 3, 6]
y2 = [1, 3, 2, 4, 5]

plt.subplot(2, 1, 1) # 2 rows, 1 column, 1st subplot
plt.plot(x, y1, color='red', marker='o', linestyle='--', linewidth=1, markersize=5, label='Line Chart')
plt.title('Line Chart')

plt.subplot(2, 1, 2) # 2 rows, 1 column, 2nd subplot
plt.bar(x, y2, color='blue', label='Bar Chart')
plt.title('Bar Chart')

plt.tight_layout()
plt.show()
```

```python
# subplots using line chart and bar chart

x = [1, 2, 3, 4, 5]
y1 = [2, 50, 5, 3, 6]
y2 = [1, 3, 2, 4, 5]

plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st subplot
plt.plot(x, y1, color='red', marker='o', linestyle='--', linewidth=1, markersize=5, label='Line Chart')
plt.title('Line Chart')

plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd subplot
plt.bar(x, y2, color='blue', label='Bar Chart')
plt.title('Bar Chart')

plt.tight_layout()
plt.show()
```

```python
# fig, ax = plt.subplots(nrows, ncols, figsize=(width, height), dpi=dpi, facecolor=facecolor, edgecolor=edgecolor, sharex=sharex, sharey=sharey, squeeze=squeeze, subplot_kw=subplot_kw, gridspec_kw=gridspec_kw)

x = [1, 2, 3, 4, 5]
y1 = [2, 50, 5, 3, 6]
y2 = [1, 3, 2, 4, 5]

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].plot(x, y1, color='red', marker='o', linestyle='--', linewidth=1, markersize=5, label='Line Chart')
ax[0].set_title('Line Chart')

ax[1].bar(x, y2, color='blue', label='Bar Chart')
ax[1].set_title('Bar Chart')

plt.suptitle('Comparison of Line Chart and Bar Chart', fontsize=16)
plt.tight_layout()
plt.show()
```
