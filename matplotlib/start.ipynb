```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 50, 5, 3, 6]

plt.plot(x,y)
plt.show()
```

```python
day = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday']
sales = [200, 100, 250, 5000, 1100]

# plt.bar(day, sales)
plt.plot(day, sales)
plt.title('Sales per day')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.show()
```
